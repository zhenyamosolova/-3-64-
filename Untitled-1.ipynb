## 1. Текстовое описание выбранного набора данных

**Название датасета:** Iris (Ирисы Фишера)

**Источник данных:** Scikit-learn library

**Краткое описание:**
Датасет Iris (Ирисы Фишера) — это классический набор данных для задач классификации, 
опубликованный британским статистиком Рональдом Фишером в 1936 году. 
Данные содержат измерения трёх видов ирисов:
- Iris setosa (Ирис щетинистый)
- Iris versicolor (Ирис разноцветный)
- Iris virginica (Ирис виргинский)

**Характеристики:**
- 150 образцов (по 50 каждого вида)
- 4 признака: длина и ширина чашелистика, длина и ширина лепестка
- Нет пропущенных значений
- Полностью сбалансированный датасет

**Почему выбран этот датасет:**
- Небольшой размер (идеально для лабораторной)
- Нет пропусков в данных
- Хорошо изучен и документирован
- Наглядно демонстрирует основные методы анализа

```python
# Импорт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Настройка отображения графиков
%matplotlib inline
sns.set_style("whitegrid")
plt.rcParams['figure.figsize'] = (12, 8)

# Загрузка данных
iris = load_iris()

# Создание DataFrame (таблицы с данными)
df = pd.DataFrame(
    data=iris.data,
    columns=iris.feature_names
)

# Добавление целевой переменной (код вида: 0,1,2)
df['target'] = iris.target

# Добавление названий видов
df['species'] = df['target'].map({
    0: 'setosa',
    1: 'versicolor', 
    2: 'virginica'
})

# Проверка, что всё загрузилось
print("Данные успешно загружены!")
print(f"Размер датасета: {df.shape[0]} строк, {df.shape[1]} столбцов")
print("\nПервые 3 строки:")
print(df.head(3))
```

  * Ширина лепестка: от 0.1 до 2.5 см

## 3. Визуальное исследование датасета

В этом разделе мы построим графики, чтобы увидеть:
- Как распределены значения каждого признака
- Есть ли различия между видами ирисов
- Какие признаки лучше всего разделяют классы

```python
# 3.1 Гистограммы распределения всех признаков
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
axes = axes.flatten()

for i, feature in enumerate(iris.feature_names):
    axes[i].hist(df[feature], bins=20, edgecolor='black', alpha=0.7, color='skyblue')
    axes[i].set_title(f'Распределение: {feature}', fontsize=12, fontweight='bold')
    axes[i].set_xlabel(feature)
    axes[i].set_ylabel('Частота')
    # Добавим линии среднего и медианы
    axes[i].axvline(df[feature].mean(), color='red', linestyle='--', linewidth=2, 
                   label=f'Среднее: {df[feature].mean():.2f}')
    axes[i].axvline(df[feature].median(), color='green', linestyle=':', linewidth=2, 
                   label=f'Медиана: {df[feature].median():.2f}')
    axes[i].legend()
    axes[i].grid(True, alpha=0.3)

plt.suptitle('Рисунок 1. Гистограммы распределения признаков', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

**Анализ гистограмм:**

- **Длина чашелистика (sepal length):** Имеет два пика (бимодальное распределение), что указывает на наличие двух разных групп в данных

- **Длина лепестка (petal length):** Явно видно разделение на две группы — маленькие значения (setosa) и большие (versicolor + virginica)
- **Ширина лепестка (petal width):** Аналогично длине лепестка, чётко видно два кластера

Среднее и медиана близки друг к другу, что говорит об отсутствии сильных выбросов.

## 4. Информация о корреляции признаков

**Что такое корреляция?**
Корреляция показывает силу связи между двумя признаками:
- **+1** — идеальная прямая связь (оба признака растут вместе)
- **0** — нет связи
- **-1** — идеальная обратная связь (один растет, другой падает)

```python
# 4.1 Создание матрицы корреляций
# Выбираем только числовые столбцы
numeric_cols = iris.feature_names + ['target']
corr_matrix = df[numeric_cols].corr()

print("=" * 60)
print("МАТРИЦА КОРРЕЛЯЦИЙ")
print("=" * 60)
print(corr_matrix.round(3))
print()

# Анализ корреляций с целевой переменной (видом)
print("=" * 60)
print("КОРРЕЛЯЦИЯ ПРИЗНАКОВ С ВИДОМ ИРИСА")
print("=" * 60)
target_corr = corr_matrix['target'].drop('target').sort_values(ascending=False)
for feature, corr in target_corr.items():
    # Определяем силу связи
    if abs(corr) > 0.7:
        strength = "🔴 очень сильная"
    elif abs(corr) > 0.5:
        strength = "🟠 сильная"
    elif abs(corr) > 0.3:
        strength = "🟡 средняя"
    else:
        strength = "⚪ слабая"
    
    direction = "положительная" if corr > 0 else "отрицательная"
    print(f"{feature}: {corr:.3f} - {strength} {direction}")
```

```python
# 4.2 Визуализация корреляций в виде тепловой карты
plt.figure(figsize=(10, 8))

# Создаем маску для верхнего треугольника (чтобы не дублировать информацию)
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Рисуем тепловую карту
sns.heatmap(
    corr_matrix,
    annot=True,           # Показывать значения
    fmt='.3f',            # Формат чисел
    cmap='RdBu_r',        # Цветовая схема (красный - положительная, синий - отрицательная)
    center=0,             # Центр цветовой шкалы
    square=True,          # Квадратные ячейки
    linewidths=1,         # Линии между ячейками
    cbar_kws={"shrink": 0.8, "label": "Коэффициент корреляции"},
    mask=mask,            # Скрываем верхний треугольник
    annot_kws={"size": 12}  # Размер цифр
)

plt.title('Рисунок 5. Тепловая карта корреляций признаков', fontsize=14, fontweight='bold')
plt.xticks(rotation=45, ha='right')
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
# 4.3 Поиск сильных корреляций между признаками
print("=" * 60)
print("СИЛЬНЫЕ КОРРЕЛЯЦИИ МЕЖДУ ПРИЗНАКАМИ (|r| > 0.7)")
print("=" * 60)

strong_correlations = []
for i in range(len(numeric_cols)):
    for j in range(i+1, len(numeric_cols)):
        corr = corr_matrix.iloc[i, j]
        if abs(corr) > 0.7:
            feature1 = numeric_cols[i]
            feature2 = numeric_cols[j]
            if feature1 != 'target' and feature2 != 'target':  # Не учитываем целевую переменную
                strong_correlations.append((feature1, feature2, corr))
                if corr > 0:
                    symbol = "⬆️"
                else:
                    symbol = "⬇️"
                print(f"{symbol} {feature1} ↔ {feature2}: {corr:.3f}")

print("\n" + "=" * 60)
print("ИНТЕРПРЕТАЦИЯ")
print("=" * 60)
print("""
Самая сильная связь - между длиной и шириной лепестка (0.96)!
Это значит, что у ирисов с длинными лепестками они почти всегда широкие.

Интересно, что длина чашелистика тоже сильно связана с параметрами лепестка,
но слабо связана с шириной чашелистика.
""")
```

```python
# 4.4 Детальный анализ самых сильных корреляций
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# График 1: Длина лепестка vs Ширина лепестка (самая сильная корреляция)
sns.scatterplot(data=df, x='petal length (cm)', y='petal width (cm)', 
                hue='species', palette='Set1', s=100, alpha=0.7, ax=axes[0,0])
axes[0,0].set_title(f'Длина ↔ Ширина лепестка\nr = {corr_matrix.loc["petal length (cm)", "petal width (cm)"]:.3f}', 
                    fontweight='bold')
axes[0,0].set_xlabel('Длина лепестка (см)')
axes[0,0].set_ylabel('Ширина лепестка (см)')
axes[0,0].grid(True, alpha=0.3)
# Добавляем линию регрессии
sns.regplot(data=df, x='petal length (cm)', y='petal width (cm)', 
            scatter=False, color='red', ax=axes[0,0])

# График 2: Длина чашелистика vs Длина лепестка
sns.scatterplot(data=df, x='sepal length (cm)', y='petal length (cm)', 
                hue='species', palette='Set1', s=100, alpha=0.7, ax=axes[0,1])
axes[0,1].set_title(f'Длина чашелистика ↔ Длина лепестка\nr = {corr_matrix.loc["sepal length (cm)", "petal length (cm)"]:.3f}', 
                    fontweight='bold')
axes[0,1].set_xlabel('Длина чашелистика (см)')
axes[0,1].set_ylabel('Длина лепестка (см)')
axes[0,1].grid(True, alpha=0.3)
sns.regplot(data=df, x='sepal length (cm)', y='petal length (cm)', 
            scatter=False, color='red', ax=axes[0,1])

# График 3: Длина лепестка vs Вид (boxplot для наглядности)
sns.boxplot(data=df, x='species', y='petal length (cm)', palette='Set2', ax=axes[1,0])
axes[1,0].set_title(f'Длина лепестка по видам\nКорреляция с видом: {target_corr["petal length (cm)"]:.3f}', 
                    fontweight='bold')
axes[1,0].set_xlabel('Вид ириса')
axes[1,0].set_ylabel('Длина лепестка (см)')
axes[1,0].grid(True, axis='y', alpha=0.3)

# График 4: Ширина лепестка vs Вид
sns.boxplot(data=df, x='species', y='petal width (cm)', palette='Set2', ax=axes[1,1])
axes[1,1].set_title(f'Ширина лепестка по видам\nКорреляция с видом: {target_corr["petal width (cm)"]:.3f}', 
                    fontweight='bold')
axes[1,1].set_xlabel('Вид ириса')
axes[1,1].set_ylabel('Ширина лепестка (см)')
axes[1,1].grid(True, axis='y', alpha=0.3)

plt.suptitle('Рисунок 6. Детальный анализ наиболее значимых корреляций', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

### **Основные выводы из корреляционного анализа**

#### **1. Корреляция с целевой переменной (видом ириса):**

| Признак | Корреляция | Сила связи |
|---------|------------|------------|
| petal width (ширина лепестка) | 0.96 | 🔴 Очень сильная |
| petal length (длина лепестка) | 0.95 | 🔴 Очень сильная |
| sepal length (длина чашелистика) | 0.78 | 🟠 Сильная |

#### **2. Корреляции между признаками:**

- **sepal length ↔ petal length: 0.87** — сильная связь

#### **3. Практические выводы:**

✅ **Для классификации** достаточно использовать только длину и ширину лепестка — они дают 95-96% информации о виде

✅ **Мультиколлинеарность:** длина и ширина лепестка очень сильно связаны, поэтому в моделях машинного обучения можно использовать только один из них

✅ **Ширина чашелистика** — наименее информативный признак, он даже отрицательно коррелирует с видом
